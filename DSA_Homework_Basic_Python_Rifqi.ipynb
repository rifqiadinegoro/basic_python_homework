### 2. Unique app categories that cover in this dataset

```python
df.iloc[10472,1:] = df.iloc[10472,1:].shift(1)
```

```python
df.drop_duplicates(subset='App', inplace=True)
```

```python
df['Installs'] = df['Installs'].apply(lambda x: x.replace('+', '') if '+' in str(x) else x)
df['Installs'] = df['Installs'].apply(lambda x: x.replace(',', '') if ',' in str(x) else x)

df['Installs'] = df['Installs'].astype(int)
df['Rating'] = df['Rating'].astype(float)
df['Reviews'] = df['Reviews'].astype(int)
```

```python
unique_cat= df["Category"].nunique() 
print(unique_cat)
```

```python
unique_gen = pd.get_dummies(df['Genres'].str.split(';').apply(pd.Series).stack()).sum(level=0)
len(unique_gen.columns)
```

### 3. Top 3 apps that have highest number of users installed/download among all categories

```python
app_ins = df.sort_values('Installs',ascending=False)
app_ins.loc[:,['App','Category','Installs']].head(3)
```

### 4. Top 3 apps that have highest number of users installed/download among the Business, Communication, Game, Dating, and Music category

```python
list_of_cat = ["BUSINESS", "COMMUNICATION", "GAME", "DATING", "MUSIC"]
data_cat = df[df['Category'].isin(list_of_cat)]
top_ins = data_cat.groupby(['Category'])
top_ins = top_ins.apply(lambda x: x.sort_values(["Installs"],ascending=False))
top_ins = top_ins[['App','Category','Installs']].reset_index(drop=True)
top_ins.groupby(['Category']).head(3)
```

### 5. Top 3 apps that have highest rating among the Business, Communication, Game, Dating, and Music category

```python
top_rat = data_cat.groupby(['Category'])
top_rat = top_rat.apply(lambda x: x.sort_values(["Rating"],ascending=False))
top_rat = top_rat[['App','Category','Rating']].reset_index(drop=True)
top_rat.groupby(['Category']).head(3)
```

### 6. Top 3 apps for the following categories in Installs and Rating criteria

```python
ins_rat = data_cat.groupby(['Category'])
ins_rat = ins_rat.apply(lambda x: x.sort_values(["Installs","Rating"],ascending=False))
ins_rat = ins_rat[['App','Category','Installs','Rating']].reset_index(drop=True)
ins_rat.groupby(['Category']).head(3)
```
